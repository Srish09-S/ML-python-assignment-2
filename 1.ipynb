```python
import pandas as pd
import numpy as np

data = {
    'creatures': ['Eagles', 'Eagles', 'Ducks', 'Herons', 'Herons',
                  'Eagles', 'Ducks', 'Eagles', 'Herons', 'Herons', 'Eagles'],
    
    'weight': [2.0, 5, 3.0, np.nan, 7, 2.5, 6, np.nan, 9, 3, 2.0],
    
    'frequency': [1, 3, 2, 5, 2, 3, 1, 3, 4, 2, 1],
    
    'status': ['no', 'no', 'yes', np.nan, 'yes', 'yes', 'yes', 'no', 'yes', 'yes', 'no'],
    
    'height': [1.2, 2.5, 1.8, 3.0, 2.2, 1.9, 2.6, 1.4, 2.8, 2.3, 1.1],
    
    'region': ['North', 'South', 'East', 'West', 'North', 'South', 'East',
               'West', 'North', 'East', 'South'],
    
    'danger_level': ['low', 'medium', 'high', 'low', 'high', 'medium',
                     'high', 'low', 'medium', 'high', 'low']
}

labels = ['x', 'y', 'z', 'w', 'v', 'u', 't', 's', 'r', 'q', 'p']

df = pd.DataFrame(data, index=labels)

print(df)
```

```python
filtered_df = df[df['frequency'] < 4]

print(filtered_df)
```

```python
df[df['frequency'] < 4]
```

```python
nan_rows = df[df['weight'].isna() | df['status'].isna()]

print(nan_rows)
```

```python
df[df['weight'].isna() | df['status'].isna()]
```

```python
for col in df.columns:
    mode_value = df[col].mode()[0]
    df[col].fillna(mode_value, inplace=True)

print(df)
```

```python
df = df.apply(lambda col: col.fillna(col.mode()[0]))
```

```python
total_visits = df[df['creatures'] == 'Herons']['frequency'].sum()

print(total_visits)
```
