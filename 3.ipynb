```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Housing.csv")
```

```python
df.groupby('bedrooms')['price'].mean().sort_values(ascending=False)
```

```python
df.groupby('bedrooms')['price'].mean().sort_values(ascending=False).head(2)
```

```python
df['PricePerSqFt'] = df['price'] / df['area']
df.groupby('stories')['PricePerSqFt'].mean()
```

```python
df.groupby('parking').size().sort_values()
```

```python
max_prices = df.groupby('hotwaterheating')['price'].max()

print(max_prices)

print(df[df['price'] == max_prices.iloc[0]])
print(df[df['price'] == max_prices.iloc[1]])
```
